```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score
import joblib
```

```python
# Load the dataset
file_path = r"C:\Users\anumf\ehr_anomaly_detection\EHR.csv"
df = pd.read_csv(file_path)

# View first few rows
df.head()

# View a few rows
df.head()
```

```python
# Handle missing values
df.fillna(method='ffill', inplace=True)

# Encode categorical variables
for col in df.select_dtypes(include='object').columns:
    df[col] = LabelEncoder().fit_transform(df[col])

# Normalize numerical features
scaler = StandardScaler()
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = scaler.fit_transform(df[num_cols])

df.head()
```

```python
df.ffill(inplace=True)
```

```python
#Split your dataset

from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# Train Isolation Forest model

from sklearn.ensemble import IsolationForest
import joblib

# Create and train the model
model = IsolationForest(contamination=0.01, random_state=42)
model.fit(X_train)

# Save the model
joblib.dump(model, 'isolation_forest_ehr.pkl')
```

```python
# Make predictions and evaluate

from sklearn.metrics import precision_score, recall_score, f1_score

# Predict on the test set
y_pred = model.predict(X_test)
y_pred = [0 if x == 1 else 1 for x in y_pred]  # Convert predictions: 1 â†’ normal, -1 â†’ anomaly

# Assume all test data is normal (label 0) for now
y_true = [0] * len(X_test)

# Calculate metrics
precision = precision_score(y_true, y_pred, zero_division=1)
recall = recall_score(y_true, y_pred, zero_division=1)
f1 = f1_score(y_true, y_pred, zero_division=1)

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get anomaly scores (the lower, the more likely to be anomaly)
scores = model.decision_function(X_test)

# Plot
plt.figure(figsize=(10,6))
plt.hist(scores, bins=50, edgecolor='black')
plt.title('Anomaly Score Distribution')
plt.xlabel('Anomaly Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Find the top 10 most anomalous samples
import numpy as np

# Get anomaly scores (lower = more anomalous)
scores = model.decision_function(X_test)

# Get indexes of top 10 most anomalous
top_anomalies_idx = np.argsort(scores)[:10]

# Show them
top_anomalies = X_test.iloc[top_anomalies_idx]
top_anomalies
```
