```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv(r"C:\Users\hp\Downloads\HospInfo.csv")  # update path if needed

# Step 2: Display initial info
print("Original Data Info:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())

# Step 3: Drop fully empty rows and columns
df.dropna(how='all', inplace=True)
df.dropna(axis=1, how='all', inplace=True)

# Step 4: Drop duplicate rows
df.drop_duplicates(inplace=True)

# Step 5: Strip whitespace from column names
df.columns = df.columns.str.strip()

# Step 6: Safely strip strings in object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)

# Step 7: Convert possible date columns
for col in df.columns:
    if 'date' in col.lower():
        try:
            df[col] = pd.to_datetime(df[col])
        except Exception:
            pass  # ignore conversion errors

# Step 8: Fill missing values
# Fill numerical columns with median
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

# Fill object columns with mode
for col in df.select_dtypes(include='object').columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Step 9: Save cleaned data
df.to_csv(r"C:\Users\hp\Downloads\HospInfo_Cleaned.csv", index=False)
print("\nCleaned dataset saved as 'HospInfo_Cleaned.csv'.")

# Step 10: Preview cleaned data
print("\nSample of cleaned data:")
print(df.head())
```

```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv(r"C:\Users\hp\Downloads\HospInfo.csv")  # update path if needed

# Step 2: Basic cleanup
df.dropna(how='all', inplace=True)                    # Drop empty rows
df.dropna(axis=1, how='all', inplace=True)            # Drop empty columns
df.drop_duplicates(inplace=True)                      # Remove duplicate rows
df.columns = df.columns.str.strip()                   # Trim column names

# Step 3: Strip string values safely
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].apply(lambda x: x.strip() if isinstance(x, str) else x)

# Step 4: Drop all footnote columns (Power BI doesn't need them)
df = df.loc[:, ~df.columns.str.contains("footnote", case=False)]

# Step 5: Convert specific columns
# Boolean to int
if 'Emergency Services' in df.columns:
    df['Emergency Services'] = df['Emergency Services'].astype(int)

# Convert 'Hospital overall rating' to numeric (clean non-numeric)
if 'Hospital overall rating' in df.columns:
    df['Hospital overall rating'] = pd.to_numeric(df['Hospital overall rating'], errors='coerce')

# Step 6: Fill missing values
# Fill numeric columns with median
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

# Fill object columns with mode
for col in df.select_dtypes(include='object').columns:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)

# Step 7: Save cleaned dataset for Power BI
output_path = r"C:\Users\hp\Downloads\HospInfo_Cleaned_PowerBI.csv"
df.to_csv(output_path, index=False, encoding='utf-8')

# Final check
print("Cleaned dataset saved for Power BI:")
print(output_path)
print("\n  Sample Preview:")
print(df.head())
```
