```python
# Check the rows where 'Tehsil' is missing
missing_tehsil = df[df['Tehsil'].isna()]

# Display rows where Tehsil is missing to identify the Districts with missing Tehsil values
print(missing_tehsil)

# Use the mode to fill missing 'Tehsil' values based on the most frequent 'Tehsil' for each District
df['Tehsil'] = df.groupby('District')['Tehsil'].transform(lambda x: x.fillna(x.mode()[0]))

# Check again for missing values
missing_tehsil_after = df[df['Tehsil'].isna()]

# Display if there are any remaining missing values
print(missing_tehsil_after)
```

```python
# Drop rows with any NaN values
df = df.dropna()
# Verify the changes by displaying the first few rows
print(df.head())
```
